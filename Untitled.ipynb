```python
from sklearn.datasets import load_breast_cancer

# Load the Breast Cancer Wisconsin dataset
data = load_breast_cancer()

# Split the data into features and target variables
X = data.data  # Features (attributes describing cell nuclei)
y = data.target  # Target variable (0 for malignant, 1 for benign)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split the dataset into a training set and a testing set (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Perform feature scaling (standardization)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# Apply initial feature selection using Fisher score (SelectKBest with f_classif)
num_features_to_select = 10  # You can choose the number of top features to select
selector = SelectKBest(score_func=f_classif, k=num_features_to_select)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Create a simple logistic regression model
logistic_regression_model = LogisticRegression(random_state=42)

# Fit the model to the training data
logistic_regression_model.fit(X_train_selected, y_train)

# Make predictions on the test data
y_pred_logistic = logistic_regression_model.predict(X_test_selected)

# Calculate the accuracy of the model
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_recall_curve, f1_score
import matplotlib.pyplot as plt

# Train an AdaBoost classifier on the training data
adaboost_model = AdaBoostClassifier(random_state=42)

# Use cross-validation to find the optimal number of base estimators (n_estimators)
param_grid = {'n_estimators': [50, 100, 150, 200]}
grid_search = GridSearchCV(estimator=adaboost_model, param_grid=param_grid, cv=5)
grid_search.fit(X_train_selected, y_train)
best_n_estimators = grid_search.best_params_['n_estimators']

# Tune other hyperparameters (e.g., learning rate) using cross-validation
param_grid = {'learning_rate': [0.1, 0.5, 1.0]}
grid_search = GridSearchCV(estimator=AdaBoostClassifier(n_estimators=best_n_estimators, random_state=42),
                           param_grid=param_grid, cv=5)
grid_search.fit(X_train_selected, y_train)
best_learning_rate = grid_search.best_params_['learning_rate']

# Fit the AdaBoost model with optimal hyperparameters
adaboost_model = AdaBoostClassifier(n_estimators=best_n_estimators, learning_rate=best_learning_rate, random_state=42)
adaboost_model.fit(X_train_selected, y_train)

# Visualize the feature importances in the model (if needed)
feature_importances = adaboost_model.feature_importances_
# You can plot feature importances if desired

# Evaluate the model's performance on the test set
y_pred_adaboost = adaboost_model.predict(X_test_selected)
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_adaboost)
f1_adaboost = f1_score(y_test, y_pred_adaboost)

# Print evaluation metrics and Precision-Recall Curve
print("AdaBoost Model Performance:")
print("Accuracy:", accuracy_adaboost)
print("F1-Score:", f1_adaboost)
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (AdaBoost)')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Train a Gradient Boosting Machine classifier on the training data
gbm_model = GradientBoostingClassifier(random_state=42)

# Use cross-validation to find optimal hyperparameters
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.1, 0.5]
}
grid_search = GridSearchCV(estimator=gbm_model, param_grid=param_grid, cv=5)
grid_search.fit(X_train_selected, y_train)
best_n_estimators = grid_search.best_params_['n_estimators']
best_max_depth = grid_search.best_params_['max_depth']
best_learning_rate = grid_search.best_params_['learning_rate']

# Fit the GBM model with optimal hyperparameters
gbm_model = GradientBoostingClassifier(n_estimators=best_n_estimators, max_depth=best_max_depth,
                                      learning_rate=best_learning_rate, random_state=42)
gbm_model.fit(X_train_selected, y_train)

# Visualize the feature importances in the model (if needed)
feature_importances = gbm_model.feature_importances_
# You can plot feature importances if desired

# Evaluate the GBM model's performance on the test set
y_pred_gbm = gbm_model.predict(X_test_selected)
accuracy_gbm = accuracy_score(y_test, y_pred_gbm)
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_gbm)
f1_gbm = f1_score(y_test, y_pred_gbm)

# Print evaluation metrics and Precision-Recall Curve
print("GBM Model Performance:")
print("Accuracy:", accuracy_gbm)
print("F1-Score:", f1_gbm)
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (GBM)')
plt.show()
```

```python
# Compare the performance of the models
print("Model Performance Comparison:")
print("AdaBoost - Accuracy: {:.2f}, F1-Score: {:.2f}".format(accuracy_adaboost, f1_adaboost))
print("GBM - Accuracy: {:.2f}, F1-Score: {:.2f}".format(accuracy_gbm, f1_gbm))
print("Logistic Regression - Accuracy: {:.2f}".format(accuracy_logistic))

# Summarize the results and provide insights
# You can compare the models and provide insights on which one performed better and why
# Discuss the impact of hyperparameter tuning on model performance
```
